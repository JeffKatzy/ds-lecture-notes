```python
def convert_to_int(text):
    no_commas = str(text).replace(',', '')
    no_dollars = no_commas[1:]
    return no_dollars
```

```python
trimmed_sixers['2018-2019 SALARY'][2][1:]
```

```python
trimmed_sixers['int_salary'] = trimmed_sixers['2018-2019 SALARY'].apply(convert_to_int)
```
